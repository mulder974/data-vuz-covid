```python
# Plot deaths per million for one country. Find Google's mobility for that country. Determine if lockdowns for that country have influenced 
# deaths per million (changed the shape of the curve). Remember, you have have to create a lagged variable to account for time from
# infection to death (median time ~ 15 days). Therefore, any change in policy ("lockdown") would take ~15 days to see an effect
```

```python
mobility = pd.read_csv('https://raw.githubusercontent.com/mulder974/data-vuz-covid/main/data_mobility_fr.csv')

querry = engine.execute("SELECT location, date, new_deaths_per_million FROM data WHERE location = 'France';").fetchall()

df4 = pd.DataFrame(data = querry)

dic= {0:"Pays",1:"date",2:"new_deaths_per_million"}

df4 = df4.rename(columns = dic, inplace = False)
# df4['date'] = pd.to_datetime(df4['date'])


df4
```

```python
mask = (mobility['place_id'] == 'ChIJMVd4MymgVA0R99lHx5Y__Ws')
mobility = mobility.loc[mask]

mobility
```

```python
df_combined = pd.merge(df4, mobility,how='left', on='date')
df_combined['date']= pd.to_datetime(df_combined['date'])
```

```python
df_combined = df_combined.dropna(how='all')
df_combined
```

```python
start_date = '2020-02-15'
end_date = '2020-12-31'

mask = (df_combined['date'] > start_date) & (df_combined['date'] <= end_date)

df_combined = df_combined.loc[mask]

col = df_combined.loc[: , "retail_and_recreation_percent_change_from_baseline":"residential_percent_change_from_baseline"]
df_combined['mean_mobility_score'] = col.mean(axis=1)

df_combined.drop(df_combined.iloc[:, 4:12], axis = 1, inplace = True)
df_combined.drop(df_combined.iloc[:, 4:8], axis = 1, inplace = True)


df_combined
```

```python
# Cretation d'une fonction pour la colonne à -14j  
```

```python
def buildLaggedFeatures(s,lag=14,dropna=True):

  if type(s) is pd.DataFrame:
      new_dict={}
      for col_name in s:
          new_dict[col_name]=s[col_name]
          # create lagged Series
          for l in range(1,lag+1):
              new_dict['%s_lag%d' %(col_name,l)]=s[col_name].shift(l)
      res=pd.DataFrame(new_dict,index=s.index)

  elif type(s) is pd.Series:
      the_range=range(lag+1)
      res=pd.concat([s.shift(i) for i in the_range],axis=1)
      res.columns=['lag_%d' %i for i in the_range]
  else:
      print('Only works for DataFrame or Series')
      return None
  if dropna:
      return res.dropna()
  else:
      return res 
```

```python
lag = buildLaggedFeatures(df_combined['new_deaths_per_million'])
```

```python
df_combined['lag_death_14days'] = lag['lag_14']
```

```python
df_combined.drop(df_combined.iloc[:, 3:5], axis = 1, inplace = True)
```

```python
df_combined['moyenne_7j_lag_new_death'] = df_combined["lag_death_14days"].rolling(7).mean()
df_combined['moyenne_7j_mobility'] = df_combined["mean_mobility_score"].rolling(7).mean()


df_combined
```

```python


fig, ax = plt.subplots()

ax1 = sns.lineplot(x="date", y="moyenne_7j_lag_new_death", data=df_combined, ax= ax)
ax2 = sns.lineplot(x="date", y="moyenne_7j_mobility", data=df_combined, ax =ax)
fig.set_size_inches(18.5, 10.5)


```
